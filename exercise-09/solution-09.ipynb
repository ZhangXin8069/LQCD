# Exercise 09

Simulation of 4d SU(2) pure gauge theory using heatbath and over-relaxation updates.

A c++ code to simulate this theory is available in the folder 'code', and the data from which these plots are made are in 'code/results'

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = (20,10)
```

```python
# Starting from a "cold" (all U = 1, plaquette = 1)
# or "hot" (all U random, plaquette ~ 0) initial
# configuration of SU(2) gauge links,
# the plaquette thermalises to the same value:

cold = np.loadtxt("code/results/thermL8_b2.3_1HB_0OR_coldstart.txt")
hot = np.loadtxt("code/results/thermL8_b2.3_1HB_0OR_hotstart.txt")
plt.figure()
plt.plot(cold[:,0], label="cold start")
plt.plot(hot[:,0], label="hot start")
plt.ylabel('plaquette')
plt.xlabel('MC updates')
plt.legend()
plt.title('Thermalisation of plaquette for L=8, beta=2.3, heatbath updates')
plt.show()
```

```python
# Different observables can have very different integrated autocorrelation times:

hb1_or0 = np.loadtxt("code/results/L6_b3.2_HB1_OR0.txt")
plt.figure()
plt.errorbar(hb1_or0[:,0], hb1_or0[:,3], yerr=hb1_or0[:,4], label="Update: 1 HB")
plt.ylabel('tau_int')
plt.xlabel('Wilson Loop size L')
plt.legend()
plt.title('tau_int of LxL Wilson loop on 6^4 lattice, beta=3.2, heatbath updates')
plt.show()
```
