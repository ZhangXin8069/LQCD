Montecarlo simulation of the 4d Ising Model using the cluster algorithm.

A c++ code to simulate the 4d Ising model using metropolis and cluster updates is available in the folder 'code', and the data from which these plots are made are in 'code/results'

```python
import matplotlib.pyplot as plt
import numpy as np

Lvalues = [4, 6, 8, 12]

def plot_data(my_plot, data_index):
    for L in Lvalues:
        str_lbl = "L=" + str(L)
        my_plot.errorbar(dat[L][:,0], dat[L][:,data_index], yerr = dat[L][:,data_index+1], marker='.', label=str_lbl)
    my_plot.axvline(x=0.1496947, label="critical beta", linestyle='--', color="grey")
    my_plot.xlabel('beta')
    my_plot.legend()
    my_plot.rcParams["figure.figsize"] = (20,10)

dat = {}
for L in Lvalues:
    dat[L] = np.loadtxt("code/results/cL" + str(L) + ".txt")
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.figure()
plot_data(plt, 1)
plt.ylabel('E/N')
plt.title('Energy vs Beta')
plt.show()
```

```python
plt.figure()
plot_data(plt, 5)
plt.ylabel('|M|/N')
plt.title('Magnetisation vs Beta')
plt.show()
```
