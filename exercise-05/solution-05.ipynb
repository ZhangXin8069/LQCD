Montecarlo simulation of the 4d Ising Model using the cluster algorithm.

A c++ code to simulate the 4d Ising model using metropolis and cluster updates is available in the folder 'code', and the data from which these plots are made are in 'code/results'

```python
import matplotlib.pyplot as plt
import numpy as np

Lvalues = [4, 6, 8, 10]

def plot_data(my_plot, data_index):
    for L in Lvalues:
        str_lbl = "L=" + str(L)
        my_plot.plot(dat[L][:,0], dat[L][:,data_index], marker='.', label=str_lbl)
    my_plot.axvline(x=0.1496947, label="critical beta", linestyle='--', color="grey")
    my_plot.xlabel('beta')
    my_plot.legend()
    my_plot.rcParams["figure.figsize"] = (20,10)

dat = {}
for L in Lvalues:
    V = np.power(L,4)
    raw_data = np.loadtxt("code/results/cL" + str(L) + ".txt")
    #construct specific heat and magnetic susceptibility from averages of  E, E^2, M, M^2:
    raw_data[:,3] = V*raw_data[:,0]*raw_data[:,0]*(raw_data[:,3] - raw_data[:,1]*raw_data[:,1])
    raw_data[:,7] = V*(raw_data[:,7] - raw_data[:,5]*raw_data[:,5])
    dat[L] = raw_data
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.figure()
plot_data(plt, 1)
plt.ylabel('E/N')
plt.title('Energy vs Beta')
plt.show()
```

```python
plt.figure()
plot_data(plt, 5)
plt.ylabel('|M|/N')
plt.title('Magnetisation vs Beta')
plt.show()
```

```python
plt.figure()
plot_data(plt, 3)
plt.ylabel('(beta^2/N)*[<E^2> - <E>^2]')
plt.title('Specific Heat vs Beta')
plt.show()
```

```python
plt.figure()
plot_data(plt, 7)
plt.ylabel('(1/N)*[<M^2> - <M>^2]')
plt.title('Magnetic Susceptibility vs Beta')
plt.show()
```
